## Score Keeping

```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline
plt.style.use('fivethirtyeight')
sns.set_context("notebook")
import warnings
warnings.filterwarnings('ignore')
```

## Midterm 2 Results

```python
scores = {
    'mechanical': sum([500]),
    'chemical': sum([500, -250, -250, 250, -250, -250, 1000, 2000]),
    'civil': sum([250, -250]),
    'mineral': sum([-1000]),
    'materials': sum([-250, 250, -500, 1000, 1000]),
    'industrial': sum([]),
    'track1': sum([250, 250, -250, -250, 1000, 2000])
}
scores = pd.DataFrame(data=scores, index=[0])

plt.figure(figsize=(15, 5))
plt.title('Midterm 2 Review Jeopardy Scores', fontsize=26)
ax = sns.barplot(x=scores.columns, y=scores.iloc[0])
ax.xaxis.set_tick_params(labelsize=16)
ax.yaxis.set_tick_params(labelsize=16)
ax.set_ylabel('Scores', fontsize=24);
```
